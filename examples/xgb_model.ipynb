```python
import pandas as pd

# Load the dataset
dataset = pd.read_csv('C:/Users/B62280/Documents/Finance/data/aapl_ohlc.csv')
dataset

```

```python
import numpy as np
rand_array = np.random.randint(2, size=len(dataset))

# Add the random array as a new column to the DataFrame
dataset['isTraded'] = rand_array
dataset.drop('Adj Close', axis=1, inplace=True)
```

```python
dataset = dataset.rename(columns={'Date': 'timestamp'})
```

```python
dataset
```

```python
window = 3

# Create lag features for open, high, low, close, and volume columns
for i in range(1, window):
    dataset[f'Open_lag_{i}'] = dataset['Open'].shift(i)
    dataset[f'High_lag_{i}'] = dataset['High'].shift(i)
    dataset[f'Low_lag_{i}'] = dataset['Low'].shift(i)
    dataset[f'Close_lag_{i}'] = dataset['Close'].shift(i)
    dataset[f'Volume_lag_{i}'] = dataset['Volume'].shift(i)

dataset
```

```python
# Drop the timestamp column and the first 2 rows that contain missing values
dataset = dataset.drop(['timestamp'], axis=1)
dataset = dataset.dropna()

# Split the dataset into input features and target variable
X = dataset.drop('isTraded', axis=1)
y = dataset['isTraded']
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
import xgboost as xgb

# Create an instance of the XGBoost classifier
xgb_clf = xgb.XGBClassifier()

# Fit the XGBoost classifier on the training data
xgb_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Predict the target variable on the testing data
y_pred = xgb_clf.predict(X_test)

# Calculate the accuracy score
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```
